```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
poke['Name'] = encoder.fit_transform(poke['Name'])
poke['Type'] = encoder.fit_transform(poke['Type'])
poke['Species'] = encoder.fit_transform(poke['Species'])
poke['Abilities'] = encoder.fit_transform(poke['Abilities'])
poke['Gender'] = encoder.fit_transform(poke['Gender'])
poke['Growth Rate'] = encoder.fit_transform(poke['Growth Rate'])
```
